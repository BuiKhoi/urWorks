```python
import requests
import os
import json
import numpy as np
```

```python
finding_dir = './data/finding/'
hiring_dir = './data/hiring/'

hiring_firebase = 'https://hackathon-2020-9448d.firebaseio.com/data_matrix/hiring'
finding_firebase = 'https://hackathon-2020-9448d.firebaseio.com/data_matrix/finding'
```

```python
label_dict = {'ban_hang': 0, 'bao_ve': 1, 'bep': 2, 'cong_nhan': 3, 'giao_vien': 4, 'giup_viec': 5, 'huong_dan_vien_du_lich': 6, 'ke_toan': 7, 'ky_su': 8, 'lai_xe': 9, 'nhan_vien': 10, 'pha_che': 11, 'phien_dich': 12, 'phuc_vu': 13, 'quan_ly': 14, 'sale_online': 15, 'tho_lap_may': 16, 'tho_xay': 17, 'thoi_vu': 18, 'tro_ly': 19, 'y_te': 20}
```

```python
def post_to_firebase(base_link, userid, content):
    r = requests.put(base_link + '/' + userid + '.json', json.dumps(content))
    return r
```

```python
def get_finding_matrix(labels):
    label = [0] * (len(label_dict) + 1)
    for l in labels[:-1]:
        label[label_dict[l]] = 1
    label[-1] = labels[-1]
    strr = ''
    for l in label[:-1]:
        strr += str(l)
    strr += 'x' + label[-1]
    return strr

def get_hiring_matrix(labels):
    label = [0] * len(label_dict)
    for l in labels:
        if l != '':
            label[label_dict[l]] = 1
    strr = ''
    for l in label:
        strr += str(l)
    return strr
```

```python
def get_index_scale_1000(index):
    if index < 10:
        return '000' + str(index)
    elif index < 100:
        return '00' + str(index)
    elif index < 1000:
        return '0' + str(index)
    else:
        return str(index)
```

```python
count = 0
for label_file in os.listdir(finding_dir):
    if count < 0:
        count += 1
        continue
    if '_label' in label_file:
        with open(finding_dir + label_file, 'r') as file_read:
            label = file_read.read()
        label = label.replace(' ', '').split('\n')
        strr = get_finding_matrix(label)
        post_to_firebase(finding_firebase, 'fnd' + get_index_scale_1000(count), strr)
        count += 1
        print('Posted {}'.format(count))
#         break
```

```python
count = 0
for label_file in os.listdir(hiring_dir):
    if count < 0:
        count += 1
        continue
    if '_label' in label_file:
        rand = np.random.random()
        if rand < 0.1:
            with open(hiring_dir + label_file, 'r') as file_read:
                label = file_read.read()
            label = label.replace(' ', '').split('\n')
            print(label)
            strr = get_hiring_matrix(label)
#             print(strr)
            if strr == '000000000000000000000':
                continue
            post_to_firebase(hiring_firebase, 'hir' + get_index_scale_1000(count), strr)
            count += 1
            print('Posted {}'.format(count))
#             break
```
